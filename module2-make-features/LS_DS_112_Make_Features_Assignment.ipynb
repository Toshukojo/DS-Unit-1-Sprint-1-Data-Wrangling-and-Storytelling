# Convert the term column from string to integer.

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 100)
! wget https://resources.lendingclub.com/LoanStats_2018Q4.csv.zip

```

```python
!unzip 'LoanStats_2018Q4.csv.zip'
```

```python
df = pd.read_csv('LoanStats_2018Q4.csv', skiprows=1, skipfooter=2, engine='python')
df.head()
```

```python
def Str_To_Float(Value):
  return int(Value.strip().strip('months'))
```

```python
Str_To_Int_list= []

for cell in df['term']:
  Str_To_Int_list.append(Str_To_Float(cell))

```

```python
df['term'] = Str_To_Int_list
df.head()
df['term'].dtypes
```

# Make a column named loan_status_is_great. It should contain the integer 1 if loan_status is "Current" or "Fully Paid." Else it should contain the integer 0.

```python
 def Loan_Stat_int(value):
   if value == 'Current':
    value = 1
   elif value == 'Fully Paid':
    value = 1
   else:
    value = 0
   return value



```

```python
loan_status_is_great = []
for value in df['loan_status']:
 Loan_status_is_great = loan_status_is_great.append(Loan_Stat_int(value))


```

```python
df['Loan_status_Is_Great'] = pd.Series(loan_status_is_great)
```
